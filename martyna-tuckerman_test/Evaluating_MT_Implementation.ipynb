# Charged water-trimer:

calculations performed using eqe with mt implementation for emebedding potential.

```python
WATER = PP('water-trimer_charged.pp').read()
WATER_MT = PP('MT_water-trimer_charged.pp').read()
```

```python
mpl.rcParams['figure.figsize'] = (10,6)

distance_scalar = 1


x0 = WATER.ions[0].pos
r0 = distance_scalar*Coord(pos=[0.0,10.0,0.0], cell = WATER.cell)
#r0 = distance_scalar*WATER.ions[0].pos.d_mic(WATER.ions[1].pos)

gridpoints = 40000
d = (r0).length()*0.52918

scalar = d/gridpoints

plot_cut1D_WATER_MT = WATER_MT.plot.get_plotcut(x0=x0, r0=r0, nr=gridpoints)
plot_cut1D_WATER = WATER.plot.get_plotcut(x0=x0, r0=r0, nr=gridpoints)

xvalues = np.arange(0,gridpoints,1)


plt.plot(xvalues*scalar,plot_cut1D_WATER_MT.values, label = 'WATER_MT' )
plt.plot(xvalues*scalar,plot_cut1D_WATER.values, label = 'WATER' )
plt.plot(xvalues*scalar,np.abs(plot_cut1D_WATER.values - plot_cut1D_WATER_MT.values), label = '|WATER - WATER_MT|' )

plt.xlabel(' d/ Å')
plt.ylabel(' v(r)')
plt.legend()

plt.show()

x0, r0
```
