## Aim

The aim of the following benchmarks is to verify the choice of the a Pseudopotential for a subsystem-DFT calculation of Cs<sub>2</sub>UO<sub>2</sub>Cl<sub>4</sub> to get the best possible embedding potential for the Uranyl-Ion in the crystal surrounding.

<div id='Performed Calculations'/>
## Performed Calculations 

The following calculations were performed for Uranium in the orthorombic crystalsystem. The lattice constants were chosen from B. E. Warren et al. (*J. Am. Chem. Soc.*, 1937, 59 (12), pp 2588â€“2591):

* a = 285.2 pm
* b = 586.5 pm
* c = 494.5 pm

For the calculations the following programms were used:

* Quantum Espresso (QE)
* exciting

For both programs only gamma-point is used for sampling the first Brillouin zone. All tested Pseudopotentials were generated with scalar relativistic calculation. In the case of exciting the calaculations were performed using "dirac" relativistic describing the core electrons and "zora" relativistic for the valence electrons.

The following XC-Functionals were used:

* PZ (QE/exciting)
* PW (QE/exciting)
* PBE (QE/exciting)
* PBE-Sol (QE/exciting)
* WC (QE/exciting)
* BP (QE)

Exact List of used Pseudopotentials Functionals (USPP: rrkjus/ PAW: kjpaw/ NC: Davide Ceresoli):

### Correlation Coefficients:

The calculation of the correlation coefficents were performed using the numpy library and the reference is always the all electron calculation from exciting for the specific XC-Functional. In the following bar graphs $$ |\mathrm{log(1 - corr. coef.)}| $$ is shown for better visualisation. 

* PW:

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import math
 
objects = ('kjpaw', 'rrkjus', 'rel-kjpaw', 'rel-rrkjus')
y_pos = np.arange(len(objects))
corr = [(-1*math.log(1-0.999499448399)), (-1*math.log(1-0.999505790596)), (-1*math.log(1-0.999571543105)),(-1*math.log(1-0.999291696277))]
 
plt.bar(y_pos, corr, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('|log(1-corr. coef.)|')
 
plt.show()
```

It seems that the PBE calculations are the most reliable in the case of USPP. Even the NC PP (Davide Ceresoli) converged for all different scalars in the case of PBE.

### Convergence

Below is the convergence for the different XC-Functionals shown. It hast to be calirified that in the case of WC and the NC-PBE PP only half of the calculations are taken into account because NC is only available with scalar relativistic and WC is only available with full relativistic. In the case of PBE every calculation (NC and US) converged completely that justify the use of pbe for further inverstigations of Cs<sub>2</sub>UO<sub>2</sub>Cl<sub>4</sub>.

```python
objects = ('PW91', 'PZ', 'PBE', 'PBE-SOL', 'BP', 'NC', 'WC')
y_pos = np.arange(len(objects))
corr = [(16/18)*100, (9/18)*100,(18/18)*100,(16/18)*100,(16/18)*100,(9/9)*100,(5/9)*100]
 
plt.bar(y_pos, corr, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('total convergence / %')
 
plt.show()
```
